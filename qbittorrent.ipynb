## Modify qBittorrent Configuration for Proxy

### Subtask:
Configure the qBittorrent settings to use the provided SOCKS5 proxy details by modifying the `qBittorrent.conf` file within the `hud8U35zYG4n` cell.


**Reasoning**:
The subtask requires configuring qBittorrent with SOCKS5 proxy details by modifying the `qBittorrent.conf` file within the `hud8U35zYG4n` cell. Since I cannot directly modify an existing cell, I will generate a new `code_block` that contains the full, updated content of the `hud8U35zYG4n` cell. This new cell will include the logic to dynamically inject the SOCKS5 proxy settings into the qBittorrent configuration based on the `proxy_details` collected in the previous step, and then restart the qBittorrent server. This effectively applies the configuration.



```python
# ============================= FORM ============================= #
#@markdown <h3>üìùCREATE CONFIG FILE</h3>
connections = 450
savePath = "/content/drive/My Drive/Repack_Hitman" #@param {type:"string"}
# ============================= FORM ============================= #

import os
import shutil
import subprocess
import threading
from IPython import display

# Mount Google Drive und sicherstellen, dass der Ordner existiert
# Re-mount if not already mounted, or if this cell is run independently
from google.colab import drive
drive.mount('/content/drive')
os.makedirs(savePath, exist_ok=True)

# Remove existing config file
shutil.rmtree('/root/.config/qBittorrent', ignore_errors=True)

configPath = r"/root/.config/qBittorrent/qBittorrent.conf"

# --- Start of Proxy Configuration Logic ---
proxy_config_to_insert = ""
# Check if proxy details are provided and not empty strings
# The proxy_details dictionary is expected to be available from a previous cell.
# Example: proxy_details = {'ip': '1.2.3.4', 'port': '8080', 'username': 'user', 'password': 'pass'}
# If no proxy is desired, proxy_ip and proxy_port should be empty strings or 'none'
# and proxy_username/password should be 'none'.

# Accessing proxy_details, connections, and savePath from the global scope/notebook environment
# For this step, we assume proxy_details has been set by the previous cell.

if proxy_details['ip'] and proxy_details['port']:
    proxy_config_to_insert += "\nConnection\\ProxyType=2"
    proxy_config_to_insert += f"\nConnection\\ProxyIP={proxy_details['ip']}"
    proxy_config_to_insert += f"\nConnection\\ProxyPort={proxy_details['port']}"

    proxy_auth_enabled = "true" if proxy_details['username'] != 'none' else "false"
    proxy_config_to_insert += f"\nConnection\\ProxyAuth={proxy_auth_enabled}"

    if proxy_details['username'] != 'none':
        proxy_config_to_insert += f"\nConnection\\ProxyUsername={proxy_details['username']}"
        proxy_config_to_insert += f"\nConnection\\ProxyPassword={proxy_details['password']}"
# --- End of Proxy Configuration Logic ---

# CREATE THE CONFIG
# Base template for qBittorrent.conf
base_data_template = r"""
[AutoRun][AutoRun]
enabled=false
program=chmod -R 666 \"%D\"

[BitTorrent]
Session\CreateTorrentSubfolder=true
Session\DisableAutoTMMByDefault=false
Session\DisableAutoTMMTriggers\CategoryChanged=false
Session\DisableAutoTMMTriggers\CategorySavePathChanged=false
Session\DisableAutoTMMTriggers\DefaultSavePathChanged=true
Session\GlobalMaxSeedingMinutes=1000

[Core]
AutoDeleteAddedTorrentFile=IfAdded

[Preferences]
Downloads\DiskWriteCacheSize=64
Session\SuggestMode=true
Advanced\AnnounceToAllTrackers=true
Session\AnnounceToAllTiers=true
Bittorrent\AddTrackers=true
Bittorrent\MaxConnecs={connections}
Bittorrent\MaxConnecsPerTorrent=-1
Bittorrent\MaxRatio=4
Bittorrent\MaxRatioAction=0
Bittorrent\MaxUploads=8
Bittorrent\PeX=true
Downloads\SavePath={savePath}
Downloads\StartInPause=false
General\Locale=en
General\UseRandomPort=True
Queueing\IgnoreSlowTorrents=true
Queueing\MaxActiveDownloads=8
Queueing\MaxActiveTorrents=15
Queueing\MaxActiveUploads=8
Queueing\QueueingEnabled=true
WebUI\Port=6006
Downloads\TempPath=/root/Downloads/temp/
Downloads\TempPathEnabled=true
WebUI\Username=rclonelab
"""

# Combine the base template with the proxy configuration and format for connections and savePath
data = (base_data_template + proxy_config_to_insert).format(connections=connections, savePath=savePath).split("\n")

# SAVE THE CONFIG
os.makedirs('/root/.config/qBittorrent/', exist_ok=True)
with open(configPath, "w") as configFile:
    for line in data:
        configFile.write(line + "\n")

print("Created the config successfully!")

# INSTALL QBITTORRENT (commands from original cell, kept for consistency if re-run)
%cd {savePath}
!add-apt-repository ppa:qbittorrent-team/qbittorrent-stable -y
!apt update
!apt install qbittorrent-nox -y

# FINAL EXECUTION
def torrent_server(port):
    # Kill any existing qbittorrent-nox processes to ensure the new config is used
    subprocess.run(['pkill', '-f', 'qbittorrent-nox'])
    subprocess.Popen(['qbittorrent-nox', f'--webui-port={port}'])

if __name__ == '__main__':
    port = 9999
    thread = threading.Thread(target=torrent_server, args=(port,))
    thread.start()
    display.clear_output(wait=True)
    print("Torrent server started!")
    print(f"File saving Location: {savePath}")
    print("Open qBittorrent WebUI via port 9999 (if running locally or through Colab, use ngrok manually).")
```

```python
# ============================= FORM ============================= #
#@markdown <h3>üìùCREATE CONFIG FILE</h3>
connections = 450
savePath = "/content/drive/My Drive/Repack_Hitman" #@param {type:"string"}
# ============================= FORM ============================= #

import os
import shutil
import subprocess
import threading
from IPython import display

# Mount Google Drive und sicherstellen, dass der Ordner existiert
# Re-mount if not already mounted, or if this cell is run independently
from google.colab import drive
drive.mount('/content/drive')
os.makedirs(savePath, exist_ok=True)

# Remove existing config file
shutil.rmtree('/root/.config/qBittorrent', ignore_errors=True)

configPath = r"/root/.config/qBittorrent/qBittorrent.conf"

# --- Start of Proxy Configuration Logic ---
proxy_config_to_insert = ""

# This assumes proxy_details has been set by a previous cell.
# If proxy_details is not defined, default to no proxy settings.
if 'proxy_details' in globals() and proxy_details['ip'] and proxy_details['port']:
    proxy_config_to_insert += "\nConnection\\ProxyType=2" # SOCKS5
    proxy_config_to_insert += f"\nConnection\\ProxyIP={proxy_details['ip']}"
    proxy_config_to_insert += f"\nConnection\\ProxyPort={proxy_details['port']}"

    proxy_auth_enabled = "true" if proxy_details['username'] != 'none' else "false"
    proxy_config_to_insert += f"\nConnection\\ProxyAuth={proxy_auth_enabled}"

    if proxy_details['username'] != 'none':
        proxy_config_to_insert += f"\nConnection\\ProxyUsername={proxy_details['username']}"
        proxy_config_to_insert += f"\nConnection\\ProxyPassword={proxy_details['password']}"
# --- End of Proxy Configuration Logic ---

# CREATE THE CONFIG
# Base template for qBittorrent.conf
base_data_template = r"""
[AutoRun][AutoRun]
enabled=false
program=chmod -R 666 \"%D\"

[BitTorrent]
Session\CreateTorrentSubfolder=true
Session\DisableAutoTMMByDefault=false
Session\DisableAutoTMMTriggers\CategoryChanged=false
Session\DisableAutoTMMTriggers\CategorySavePathChanged=false
Session\DisableAutoTMMTriggers\DefaultSavePathChanged=true
Session\GlobalMaxSeedingMinutes=1000

[Core]
AutoDeleteAddedTorrentFile=IfAdded

[Preferences]
Downloads\DiskWriteCacheSize=64
Session\SuggestMode=true
Advanced\AnnounceToAllTrackers=true
Session\AnnounceToAllTiers=true
Bittorrent\AddTrackers=true
Bittorrent\MaxConnecs={connections}
Bittorrent\MaxConnecsPerTorrent=-1
Bittorrent\MaxRatio=4
Bittorrent\MaxRatioAction=0
Bittorrent\MaxUploads=8
Bittorrent\PeX=true
Downloads\SavePath={savePath}
Downloads\StartInPause=false
General\Locale=en
General\UseRandomPort=True
Queueing\IgnoreSlowTorrents=true
Queueing\MaxActiveDownloads=8
Queueing\MaxActiveTorrents=15
Queueing\MaxActiveUploads=8
Queueing\QueueingEnabled=true
WebUI\Port=6006
Downloads\TempPath=/root/Downloads/temp/
Downloads\TempPathEnabled=true
WebUI\Username=rclonelab
""".format(connections=connections, savePath=savePath)

# Combine the base template with the proxy configuration
data = (base_data_template + proxy_config_to_insert).split("\n")

# SAVE THE CONFIG
os.makedirs('/root/.config/qBittorrent/', exist_ok=True)
with open(configPath, "w") as configFile:
    for line in data:
        configFile.write(line + "\n")

print("Created the config successfully!")

# INSTALL QBITTORRENT (commands from original cell, kept for consistency if re-run)
# The user can choose to comment these out if already installed.
# %cd {savePath}
# !add-apt-repository ppa:qbittorrent-team/qbittorrent-stable -y
# !apt update
# !apt install qbittorrent-nox -y

# FINAL EXECUTION
def torrent_server(port):
    # Kill any existing qbittorrent-nox processes to ensure the new config is used
    subprocess.run(['pkill', '-f', 'qbittorrent-nox'])
    subprocess.Popen(['qbittorrent-nox', f'--webui-port={port}'])

if __name__ == '__main__':
    port = 9999
    thread = threading.Thread(target=torrent_server, args=(port,))
    thread.start()
    display.clear_output(wait=True)
    print("Torrent server started!")
    print(f"File saving Location: {savePath}")
    print("Open qBittorrent WebUI via port 9999 (if running locally or through Colab, use ngrok manually).")
```
